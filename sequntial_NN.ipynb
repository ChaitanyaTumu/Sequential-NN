This project demonstrates the implementation of a basic Artificial Neural Network (ANN) using TensorFlow and Keras to solve the classic XOR (Exclusive OR) logic gate problem.

```python
import torch
import tensorflow as tf
import numpy as np
from tensorflow import sequence_mask
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
```

### 2. Defining the Dataset and Model Architecture
We define our dataset based on the XOR logic gate truth table:
* Input `[0, 0]` -> Output `[0]`
* Input `[0, 1]` -> Output `[1]`
* Input `[1, 0]` -> Output `[1]`
* Input `[1, 1]` -> Output `[0]`

Next, we build a Sequential model. It consists of:
1. **An Input Layer** specifying the 2 input features.
2. **A Hidden Layer** with 3 neurons and a Sigmoid activation function to handle the non-linearity.
3. **An Output Layer** with 1 neuron and a Sigmoid activation function to output a probability between 0 and 1.

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([[0],[1],[1],[0]])

model=Sequential()

model.add(Dense(3,input_dim=2,activation='sigmoid'))
model.add(Dense(1,input_dim=3,activation='sigmoid'))
```

### 3. Model Compilation
Before training, we must compile the model by defining how it will learn:
* **Optimizer:** Stochastic Gradient Descent (SGD) with a relatively high learning rate of 0.3 to help the network converge quickly on this small dataset.
* **Loss Function:** Mean Squared Error (MSE), which will measure how far off our predictions are from the actual values.

```python
# Define the optimizer and set the learning rate
sgd = SGD(learning_rate=0.3)

# Compile the model with MSE loss and the SGD optimizer
model.compile(loss='mean_squared_error', optimizer=sgd)

# Output a summary of the model's architecture and parameter count
model.summary()
```

### 4. Training the Neural Network
We train the model for 1,000 epochs. Since our dataset is very small (4 samples), we pass the entire dataset in a single batch (`batch_size=4`).

After training, we use `model.predict()` to see how well our network learned the XOR logic.

```python
# Train the model on the input data
# epochs=1000: The model will iterate over the entire dataset 1,000 times
# batch_size=4: All 4 samples are processed before updating the model weights
model.fit(x, y, batch_size=4, epochs=1000, verbose=1)

# Generate predictions based on the trained weights
print("\n--- Final Predictions ---")
predictions = model.predict(x)
print(predictions)

# Optional: Print rounded predictions to show crisp 0s and 1s
# print("\n--- Rounded Predictions ---")
# print(np.round(predictions))
```
